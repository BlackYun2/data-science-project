```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
import math
```

```python
url1 = "https://www.disastercenter.com/crime/nycrime.htm"
url2 = "https://www1.nyc.gov/site/nypd/news/p00039/nypd-citywide-crime-statistics-february-2022"
url3 = "https://ucr.fbi.gov/crime-in-the-u.s/2018/crime-in-the-u.s.-2018/tables/table-8/table-8-state-cuts/new-york.xls"

```

```python
html1 = requests.get(url1)
html2 = requests.get(url2)
html3 = requests.get(url3)
```

```python
html1_text = html1.text
page1 = BeautifulSoup(html1_text, "html.parser")
html2_text = html2.text
page2 = BeautifulSoup(html2_text, "html.parser")
html3_text = html3.text
page3 = BeautifulSoup(html3_text, "html.parser")

```

```python
table = page1.find("table")
```

```python
s = []
temp = table.find("tbody").find("tbody")

temp1 = temp.findAll("tbody")
rows = temp1[2].findAll("tr")
rows[1]
```

```python
s = [row.findAll("td") for row in rows]
for item in s:
    for j in range(len(item)):
        item[j] = item[j].text
        item[j] = item[j].replace("\xa0", "")
        item[j] = item[j].replace("\n", "")
        item[j] = item[j].replace(" ", "")
s
```

```python
Year = []
Population = []
Index = []
Violent = []
Property = []
Murder = []
Forcible_Rape = []
Robbery = []
Aggravated_assault = []
Burglary = []
Larceny_Theft = []
Vehicle_Theft = []

for i in s:
    if i[0] in ["", "Year"]:
        continue
    Year.append(i[0])
    Population.append(i[1])
    Index.append(i[2])
    Violent.append(i[3])
    Property.append(i[4])
    Murder.append(i[5])
    Forcible_Rape.append(i[6])
    Robbery.append(i[7])
    Aggravated_assault.append(i[8])
    Burglary.append(i[9])
    Larceny_Theft.append(i[10])
    Vehicle_Theft.append(i[11])
Year
```

```python
df = pd.DataFrame(
    {
        "Year": Year,
        "Population": Population,
        "Index": Index,
        "Violent": Violent,
        "Property": Property,
        "Murder": Murder,
        "Forcible_Rape": Forcible_Rape,
        "Robbery": Robbery,
        "Aggravated_assault": Aggravated_assault,
        "Burglary": Burglary,
        "Larceny_Theft": Larceny_Theft,
        "Vehicle_Theft": Vehicle_Theft,
    }
)
df
```
